Inspect Model

```python
from keras.applications import VGG16
```

```python
def inspect_model(include_top=True):
    # load the VGG16 network
    print("[INFO] loading network...")
    model = VGG16(weights="imagenet", include_top=include_top)
    print("[INFO] showing layers...")

    # loop over the layers in the network and display them to the
    # console
    for (i, layer) in enumerate(model.layers):
        print("[INFO] {}\t{}".format(i, layer.__class__.__name__))
```

```python
inspect_model(include_top=True)
```

```python
inspect_model(include_top=False)
```
