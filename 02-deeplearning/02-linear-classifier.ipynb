```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
# initialize the class labels and set the seed of the pseudorandom
# number generator so we can REPRODUCE our results
labels = ["dog", "cat", "panda"]
np.random.seed(1)

# randomly initialize our weight matrix and bias vector -- in a *real* training and classification task, 
# these parameters would be *learned* by our model, but for the sake of this example, let's use random values
W = np.random.randn(3, 3072)
b = np.random.randn(3)

orig = cv2.imread("drive/MyDrive/pyimagesearch/datasets/beagle.png")
image = cv2.resize(orig, (32, 32)).flatten()

scores = W.dot(image) + b

for (label, score) in zip(labels, scores):
	print("[INFO] {}: {:.2f}".format(label, score))
 
plt.axis("off")
plt.title("Label: {}".format(labels[np.argmax(scores)]))
plt.imshow(cv2.cvtColor(orig, cv2.COLOR_BGR2RGB))
```
